```python
# for 문의 기본 구조

# for 변수 in 리스트(또는 튜플, 문자열):
#     수행할_문장1
#     수행할_문장2

#리스트,튜플, 문자열의 갯수는 len 으로 구할 수 있음. 길이가 즉 갯수가 리스트 변수가 반복되서 받아준다  
# while 은 무한 루프가 걸릴 수 있지만 리스트는 len의 갯수만큼만 반복되기 때문에 멈춘다
```

```python
# 전형적인 for 문
test_list = ['one', 'two', 'three'] 
for i in test_list: 
    print(i)
```

```python
# 다양한 for 문의 사용

a = [(1,2), (3,4), (5,6)]  #요소는 3개  # 튜플 

for (a,b) in a:    #(a,b) 대신 (a, ) 도 가능하다 튜플이기 때문에
    print(a+b)
```

```python
# for 문의 응용

#총 5명의 학생 시험 점수

scores = [90, 25, 67, 45, 80]
id = 0

for score in scores:
    id += 1 #id = id+ 1
    if score >= 60:
        print(f"id {id}, 합격입니다.")
    else:
        print(f"di: {id}, 불합격입니다.")
```

```python
#for 문과 continue 문
scores = [90, 25, 67, 45, 80]
id = 0

for score in scores:     #score 변수 
    id += 1

    if score < 60:
        continue
    print(f"id가 {id} 학생은 합격입니다.")
```

```python
# 1부터 10까지 더하는 반복문
# range 는 시작은 포함 마지막은 미포함.

add = 0 

for i in range(1,10+1): # 1부터 10까지 더하는거. 
 add = add + i
 print(add)   #이건 과정중에 더한거고

 print(add)  #이건 마지막에 더한거임
```

```python
# for와 range()를 이용한 구구단을 만들어보세요.

#출력예시는 
# 2 * 1 = 2
# 2 * 2 = 4
# 2 * 3 = 6
# ...
# 9 * 9 = 81

for a in range(2,10): #[2,3,4,....9]
    for b in range(1,10): # [1,2,3,...9]
        print(f"{a} * {b} = {a * b}")
```

```python
# zip 함수로 여러 리스트 함께 순회하기
names = ["홍길동","유재석","강호동"]
scores = [85, 92,80]

for name, score in zip(names,scores):
    print(f"{name}:{score}점")
```
