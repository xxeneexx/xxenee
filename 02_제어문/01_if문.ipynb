```python
#IF문은 왜 필요할까?
"돈이 있으면 택시를 타고 가고, 돈이 없으면 걸어간다"
```

```python
money = True #bool 자료형임

if money: print("택시를 타고 가라")  #if 다음에는 논리식으로 자리가 와야함 로직컬테스트임
else: print("걸어 가라.")
```

```python
# if문의 기본구조

# if 조건문:           #문장이 아직 안끝났기때문에 그러면 이라는뜻 들여쓰기
#     수행할_문장1      #내어쓰기
#     수행할_문장2      # 수행할 것이 여러개일때는 줄을 바꿔서 해야하고 수행할것이 하나이면 한줄로
#     ...
# else:
#     수행할_문장A
#     수행할_문장B
#     ...
```

```python
#들여쓰기 방법

# if 조건문:
#     수행할_문장1  
# 수행할_문장2 -> 줄이 안 맞으면 if문 안에서 실행이 되지 않음 
#     수행할_문장3 -> 에러 발생


money = True
if money: 
    print("타고")
    print("가라")
    print("택시를")
```

```python
money = 2000
if money >= 3000:
    print("택시를 타고 가라")
else:
    print("걸어가라")

```

```python
# 돈이 3000원 이상 있거나 카드가 있다면 택시를 타고 가고,
# 그렇지 않으면 걸어가라.

money = 2000
card = False
if money >= 3000 or card == True:
   print("택시를 타고 가라")
else:
   print("걸어가라")
 
```

```python
#만약에 주머니에 money 가 있으면 택시를 타고가고, 그렇지 않으면 걸어가라

pocket = ["paper","phone","money"]

if "money" in pocket:   
    print("택시를 타고 가라")
else:
    print("걸어가라")
```

```python
#다양한 조건을 판단하는 elif

pocket = ["paper","phone"]
card = True

if "money" in pocket:
    print("택시를 타고 가라")
elif card == True:  # elif를 사용해서 else-if 구조
    print("택시를 타고 가라")
else:
    print("걸어가라")
```

```python
# 조건부 표현식
score = 85

if score >= 60:
    result = "합격"
else:
    result = "불합격"

print(result)
```

```python
# 학점이 90이상이면 A, 
# 80이상 90미만이면 B,
# 70이상 80미만이면 C,
# 60이상 70미만이면 D,
# 60이상 F
# 출력하는 프로그램을 만드시오.

#숫자를 input 으로하면 타입이 문자, int로하면 숫자로  

score = int(input("학점을 입력하세요 : "))

if score >= 90:
    print(f"입력값은{score},당신의 학점은 :A")
elif score >= 80:
    print(f"입력값은{score},당신의 학점은 :B")
elif score >= 70:
    print(f"입력값은{score},당신의 학점은 :C")
elif score >= 60:
    print(f"입력값은{score},당신의 학점은 :D")
elif score > 60:
    print(f"입력값은{score},당신의 학점은 :F")

```
