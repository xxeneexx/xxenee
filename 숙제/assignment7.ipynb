````python
# 성적 관리 시스템
# 학생들의 성적을 입력받아 파일에 저장하고, 평균을 계산하는 종합 프로그램을 작성하세요.

# 프로그램 기능:
# 1. 학생 이름과 점수(국어, 영어, 수학)를 입력받기
# 2. 각 학생의 평균 계산 (함수 사용)
# 3. 결과를 grades.txt 파일에 저장
# 4. 저장된 파일 읽어서 전체 출력

# def calculate_average(korean, english, math):
#     """세 과목의 평균을 계산하는 함수"""
#     pass

# def save_grade(name, korean, english, math, average):
#     """성적 정보를 파일에 저장하는 함수"""
#     pass

# def display_grades():
#     """파일에서 성적을 읽어 출력하는 함수"""
#     pass
# ```

# **실행 예시:**
# ```
# 학생 이름: 김철수
# 국어 점수: 85
# 영어 점수: 90
# 수학 점수: 88
# 평균: 87.67점

# 성적이 저장되었습니다!

# 다른 학생을 추가하시겠습니까? (y/n): y

# 학생 이름: 이영희
# 국어 점수: 92
# 영어 점수: 88
# 수학 점수: 95
# 평균: 91.67점

# 성적이 저장되었습니다!

# 다른 학생을 추가하시겠습니까? (y/n): n

# ===== 전체 성적 =====
# 김철수: 국어 85, 영어 90, 수학 88 (평균: 87.67)
# 이영희: 국어 92, 영어 88, 수학 95 (평균: 91.67)
# ===================

#round 는 소수점을 반올림하는거 
def calculate_average(korean, english, math):
    """세 과목의 평균을 계산하는 함수"""
    return round((korean + english + math) / 3, 2)

def save_grade(name, korean, english, math, average):
    """성적 정보를 파일에 저장하는 함수"""
    with open("grades.txt", "a", encoding="utf-8") as f:  # as if 는 앞에있는 긴 문자를 f로 쓰겠다 
        f.write(f"{name},{korean},{english},{math},{average}\n")
#write는 객체에있는 메써드   #encoding 우리가 읽는글자를 컴퓨터가 읽게끔하는거 왜냐면 컴퓨터는 뭐라써도 0,1로만 읽어서
def display_grades():
    """파일에서 성적을 읽어 출력하는 함수"""
    print("\n===== 전체 성적 =====")
    with open("grades.txt", "r", encoding="utf-8") as f:
        for line in f:
            name, korean, english, math, average = line.strip().split(",") #체이닝 방식?
            print(f"{name}: 국어 {korean}, 영어 {english}, 수학 {math} (평균: {average})")
    print("===================")

# 학생 입력 반복
while True:
    name = input("학생 이름: ")
    korean = int(input("국어 점수: "))
    english = int(input("영어 점수: "))
    math = int(input("수학 점수: "))

    avg = calculate_average(korean, english, math)
    print(f"평균: {avg}점")

    save_grade(name, korean, english, math, avg)
    print("\n성적이 저장되었습니다!\n")

    more = input("다른 학생을 추가하시겠습니까? (y/n): ")
    if more.lower() != "y":
        break

display_grades()


````
