```python
#투두리스트를 자유롭게 개발해주세요 


# =============================
# 할 일 클래스
# =============================
class Task:
    def __init__(self, title):
        self.title = title
        self.done = False

    def complete(self):
        self.done = True


# =============================
# 투두리스트 관리 클래스
# =============================
class TodoList:
    def __init__(self):
        self.tasks = []

    # 1. 할 일 추가
    def add_task(self):
        title = input("추가할 할 일을 입력하세요: ")
        task = Task(title)
        self.tasks.append(task)
        print("할 일이 추가되었습니다.")

    # 2. 할 일 목록 조회
    def show_tasks(self):
        if not self.tasks:
            print("할 일이 없습니다.")
            return

        print("\n[할 일 목록]")
        for i, task in enumerate(self.tasks, 1):
            status = "완료" if task.done else "미완료"
            print(f"{i}. {task.title} ({status})")

    # 3. 할 일 완료 처리
    def complete_task(self):
        self.show_tasks()
        if not self.tasks:
            return

        try:
            num = int(input("완료할 할 일 번호를 입력하세요: "))
            self.tasks[num - 1].complete()
            print("완료 처리되었습니다.")
        except:
            print("잘못된 번호입니다.")

    # 4. 할 일 삭제
    def delete_task(self):
        self.show_tasks()
        if not self.tasks:
            return

        try:
            num = int(input("삭제할 할 일 번호를 입력하세요: "))
            del self.tasks[num - 1]
            print("삭제되었습니다.")
        except:
            print("잘못된 번호입니다.")


# =============================
# 실행 코드
# =============================
todo = TodoList()

while True:
    print("\n===== 투두 리스트 =====")
    print("1. 할 일 추가")
    print("2. 할 일 목록 조회")
    print("3. 할 일 완료 처리")
    print("4. 할 일 삭제")
    print("5. 종료")

    menu = input("메뉴 선택: ")

    if menu == "1":
        todo.add_task()
    elif menu == "2":
        todo.show_tasks()
    elif menu == "3":
        todo.complete_task()
    elif menu == "4":
        todo.delete_task()
    elif menu == "5":
        print("프로그램 종료")
        break
    else:
        print("올바른 메뉴를 선택하세요.")
```
