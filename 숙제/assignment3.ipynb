```python

### 문제 1-1. 딕셔너리 기본 개념
다음 질문에 답하시오.

1. 딕셔너리는 무엇으로 구성되어 있는가?
-Key와 Value가 쌍으로 이루어진 자료형
-Key: 값을 식별하는 이름으로 숫자, 문자열도 가능
-Value: Key가 가키는 실제 데이터 

구조)
a = {'key : Value'}

2. 딕셔너리와 리스트의 가장 큰 차이점은 무엇인가?

딕셔너리는 Key와 Value가 쌍을 이루고 있어 키를 알면 값을 바로 찾을 수 있다
ex)locker = {1:"책가방", 2:"운동화", 3:"도시락"}
print(locker[2]) '운동화'

리스트는 순서대로 놓아진 값을들 인덱스로 위치 확인한다 
ex) homework = ["국어", "수학", "영어"]
homework[0] → 첫 번째 숙제: 국어



3. 다음 중 딕셔너리의 Key로 사용할 수 없는 것은? (이유도 함께 설명) 
- c  key의 조건은 변경 불가능해야한다 하나의 객체만하다. 리스트는 변경 되기때문       
   - a) 숫자 (1, 2, 3)
   - b) 문자열 ("name", "age")
   - c) 리스트 ([1, 2, 3])
   - d) 튜플 ((1, 2, 3))

### 문제 1-2. 코드 읽기
다음 코드의 실행하시오

student = {'name': '김철수', 'age': 20, 'major': '컴퓨터공학'}
print(student['name'])
print(student['age'])

#
김철수
20


---

### 문제 2-1. 딕셔너리 생성 및 조작
다음 조건에 맞는 딕셔너리를 생성하고 조작하시오.


# 1. 다음 정보를 담는 딕셔너리를 만드시오
#    - 이름: 본인 이름
#    - 나이: 본인 나이
#    - 취미: 본인 취미 (리스트로 여러 개)
#    - 거주지: 본인 거주지

a ={
    'name' : 'enee' , 'age' : 32 , 'hobby': ['yoga','workout','self_development'],
'location' : 'phohang'}
print(a['name'])
print(a['age'])
print(a['hobby'])
print(a["location"])

# 2. 위 딕셔너리에 '직업' 항목을 추가하시오
a['occupation'] = 'student'
print(a['occupation'])

# 3. '취미' 항목에 새로운 취미를 추가하시오
a['hobby'].append('reading')  #-> a['hobby']=('reading') 이거도됨
print(a['hobby'])
# 4. '나이' 항목의 값을 1 증가시키시오

a["age"]+=1
print(a['age'])

# 5. 모든 key와 value를 출력하시오

# 1. 다음 정보를 담는 딕셔너리를 만드시오
#    - 이름: 본인 이름
#    - 나이: 본인 나이
#    - 취미: 본인 취미 (리스트로 여러 개)
#    - 거주지: 본인 거주지

a ={
    'name' : 'enee' , 'age' : 32 , 'hobby': ['yoga','workout','self_development'],
'location' : 'phohang'}
print(a['name'])
print(a['age'])
print(a['hobby'])
print(a["location"])

# 2. 위 딕셔너리에 '직업' 항목을 추가하시오
a['occupation'] = 'student'
print(a['occupation'])

# 3. '취미' 항목에 새로운 취미를 추가하시오
a['hobby'].append('reading')
print(a['hobby'])
# 4. '나이' 항목의 값을 1 증가시키시오

a["age"]+=1
print(a['age'])

# 5. 모든 key와 value를 출력하시오

enee
32
['yoga', 'workout', 'self_development']
phohang
student
['yoga', 'workout', 'self_development', 'reading']
33


### 문제 2-2. 딕셔너리 함수 활용
#다음 딕셔너리가 주어졌을 때, 각 요구사항을 수행하시오.


fruits = {
    'apple': 1000,
    'banana': 1500,
    'orange': 2000,
    'grape': 3000
}

# 1. 모든 과일 이름(key)을 리스트로 출력하시오
fruits.keys()
list(fruits.keys()) 
print(list(fruits.keys()))

# 2. 모든 가격(value)을 리스트로 출력하시오
fruits.values()
list(fruits.values()) 
print(list(fruits.values()))

# 3. 'banana'의 가격을 출력하시오
print(fruits['banana'])

# 4. 'melon'이 딕셔너리에 있는지 확인하시오
print('melon in fruits')

# 5. 'grape'를 pop() 함수를 사용하여 삭제하고, 삭제된 값을 출력하시오
grape = fruits.pop('grape')
print(grape)

---

### 문제 3-1. 성적 관리 프로그램
학생들의 성적을 관리하는 프로그램을 작성하시오.

# 다음 학생들의 성적 정보를 딕셔너리로 저장하시오
# 학생1: 이름=김민수, 국어=85, 영어=90, 수학=78
# 학생2: 이름=이영희, 국어=92, 영어=88, 수학=95
# 학생3: 이름=박철수, 국어=78, 영어=85, 수학=82

# 요구사항:
# 1. 각 학생의 총점과 평균을 계산하여 딕셔너리에 추가
# 2. 각 학생의 정보를 보기 좋게 출력
# 3. 평균이 가장 높은 학생 찾기


print("=== 학생 성적 관리 ===")

# 학생별 점수 딕셔너리
grade = {
    "김민수": {"국어":85, "영어":90, "수학":78},
    "이영희": {"국어":92, "영어":88, "수학":95},
    "박철수": {"국어":78, "영어":85, "수학":82}
}

# 최고 평균 저장용 변수
top_student = ""
top_avg = 0

# 김민수
total = grade["김민수"]["국어"] + grade["김민수"]["영어"] + grade["김민수"]["수학"]
avg = total / 3
print("이름: 김민수")
print("국어:", grade["김민수"]["국어"], "점, 영어:", grade["김민수"]["영어"], "점, 수학:", grade["김민수"]["수학"], "점")
print("총점:", total, "점, 평균:", "{:.2f}".format(avg), "점")
print("---")
if avg > top_avg:
    top_avg = avg
    top_student = "김민수"

# 이영희
total = grade["이영희"]["국어"] + grade["이영희"]["영어"] + grade["이영희"]["수학"]
avg = total / 3
print("이름: 이영희")
print("국어:", grade["이영희"]["국어"], "점, 영어:", grade["이영희"]["영어"], "점, 수학:", grade["이영희"]["수학"], "점")
print("총점:", total, "점, 평균:", "{:.2f}".format(avg), "점")
print("---")
if avg > top_avg:
    top_avg = avg
    top_student = "이영희"

# 박철수
total = grade["박철수"]["국어"] + grade["박철수"]["영어"] + grade["박철수"]["수학"]
avg = total / 3
print("이름: 박철수")
print("국어:", grade["박철수"]["국어"], "점, 영어:", grade["박철수"]["영어"], "점, 수학:", grade["박철수"]["수학"], "점")
print("총점:", total, "점, 평균:", "{:.2f}".format(avg), "점")
print("---")
if avg > top_avg:
    top_avg = avg
    top_student = "박철수"

# 최고 성적 학생 출력
print("최고 성적 학생:", top_student, "(평균:", "{:.2f}".format(top_avg), "점)")


# 출력 결과
=== 학생 성적 관리 ===
이름: 김민수
국어: 85 점, 영어: 90 점, 수학: 78 점
총점: 253 점, 평균: 84.33 점
---
이름: 이영희
국어: 92 점, 영어: 88 점, 수학: 95 점
총점: 275 점, 평균: 91.67 점
---
이름: 박철수
국어: 78 점, 영어: 85 점, 수학: 82 점
총점: 245 점, 평균: 81.67 점
---
최고 성적 학생: 이영희 (평균: 91.67 점)


**출력 예시:**

=== 학생 성적 관리 ===
이름: 김민수
국어: 85점, 영어: 90점, 수학: 78점
총점: 253점, 평균: 84.33점
---
이름: 이영희
국어: 92점, 영어: 88점, 수학: 95점
총점: 275점, 평균: 91.67점
---
이름: 박철수
국어: 78점, 영어: 85점, 수학: 82점
총점: 245점, 평균: 81.67점
---
최고 성적 학생: 이영희 (평균: 91.67점)

### 문제 3-2. 단어 카운터
문자열에서 각 단어가 몇 번 등장하는지 세는 프로그램을 작성하시오.

text = "python is great python is fun python is powerful"

# 요구사항:
# 1. 문자열을 공백으로 분리하여 단어 리스트 만들기
# 2. 딕셔너리를 사용하여 각 단어의 등장 횟수 세기
# 3. 결과를 출력

# 힌트: 
# - 딕셔너리에 key가 없을 때 get() 함수의 default 값 활용
# - 또는 'word' in dict_name 으로 확인


# 1. 공백으로 분리하여 단어 리스트 만들기
words = text.split(" ")

# 2. 단어 등장 횟수 세기
word_count = {}

# 단어 하나씩 확인하며 개수 세기
for w in words:
    # key가 없으면 0, 있으면 현재 값 +1
    word_count[w] = word_count.get(w, 0) + 1

# 3. 결과 출력
print("=== 단어 등장 횟수 ===")
print("python:", word_count["python"], "번")
print("is:", word_count["is"], "번")
print("great:", word_count["great"], "번")
print("fun:", word_count["fun"], "번")
print("powerful:", word_count["powerful"], "번")

#출력 결과
=== 단어 등장 횟수 ===
python: 3 번
is: 3 번
great: 1 번
fun: 1 번
powerful: 1 번


**출력 예시:**

python: 3번
is: 3번
great: 1번
fun: 1번
powerful: 1번



**출력 예시:**

python: 3번
is: 3번
great: 1번
fun: 1번
powerful: 1번


### 문제 4-1. 중첩 딕셔너리
다음과 같은 중첩 딕셔너리를 다루는 문제를 해결하시오.


school = {
    '1반': {
        '김철수': {'국어': 85, '영어': 90, '수학': 78},
        '이영희': {'국어': 92, '영어': 88, '수학': 95}
    },
    '2반': {
        '박민수': {'국어': 88, '영어': 85, '수학': 90},
        '최지영': {'국어': 95, '영어': 92, '수학': 88}
    }
}

# 요구사항:
# 1. '1반' 김철수의 수학 점수를 출력하시오
# 2. '2반' 박민수의 평균 점수를 계산하여 출력하시오
# 3. 전교에서 국어 점수가 가장 높은 학생을 찾으시오
# 4. 각 반의 평균 점수를 계산하시오

# 1. '1반' 김철수의 수학 점수 출력
print("1반 김철수의 수학 점수:", school['1반']['김철수']['수학'], "점")

# 2. '2반' 박민수 평균 점수 계산
total = school['2반']['박민수']['국어'] + school['2반']['박민수']['영어'] + school['2반']['박민수']['수학']
avg = total / 3
print("2반 박민수 평균 점수:", "{:.2f}".format(avg), "점")

# 3. 전교에서 국어 점수가 가장 높은 학생 찾기
# 학생 이름, 점수 저장용
top_student = ""
top_score = 0

# 1반 학생들
for name, scores in school['1반'].items():
    if scores['국어'] > top_score:
        top_score = scores['국어']
        top_student = name

# 2반 학생들
for name, scores in school['2반'].items():
    if scores['국어'] > top_score:
        top_score = scores['국어']
        top_student = name

print("전교 국어 1등 학생:", top_student, "(점수:", top_score, "점)")

# 4. 각 반의 평균 점수 계산
# 1반 평균
total1 = 0
count1 = 0
for scores in school['1반'].values():
    total1 += scores['국어'] + scores['영어'] + scores['수학']
    count1 += 3
avg1 = total1 / count1
print("1반 평균 점수:", "{:.2f}".format(avg1), "점")

# 2반 평균
total2 = 0
count2 = 0
for scores in school['2반'].values():
    total2 += scores['국어'] + scores['영어'] + scores['수학']
    count2 += 3
avg2 = total2 / count2
print("2반 평균 점수:", "{:.2f}".format(avg2), "점")

#출력 결과

1반 김철수의 수학 점수: 78 점
2반 박민수 평균 점수: 87.67 점
전교 국어 1등 학생: 최지영 (점수: 95 점)
1반 평균 점수: 87.75 점
2반 평균 점수: 88.25 점


### 문제 5. 다음 코드의 실행 결과를 예상하시오.

# 1)
s1 = set("Hello")
print(s1)

- {'e','H','l','o'}

# 2)
s2 = set([1, 2, 3, 2, 1, 3, 4])
print(s2)

- {1,2,3,4}

# 3)
s3 = {}
print(type(s3)
)
- <class 'dic'>

#참고 **빈 집합** 
s3 = set()
print(type(s3))  # <class 'set'>



### 문제 6. 두 반의 학생 명단 비교

A반과 B반 학생 명단이 있습니다.


b_class = {'박민수', '최지영', '정수민', '강호동', '이영희'}
a_class = {'김철수', '이영희', '박민수', '최지영', '홍길동'}

다음 질문에 답하시오:


# 1. 두 반 모두에 있는 학생 (교집합)
print("교집합 : ", a_class & b_class)
교집합 :  {'최지영', '박민수', '이영희'}

# 2. 두 반을 합친 전체 학생 (합집합)
print("합집합 : ", a_class | b_class)
합집합 :  {'최지영', '이영희', '김철수', '강호동', '홍길동', '박민수', '정수민'}

# 3. A반에만 있는 학생 (차집합)

print("차집합 : ", a_class - b_class)
차집합 :  {'홍길동', '김철수'}

# 4. B반에만 있는 학생 (차집합)
print("차집합 : ", b_class - a_class )
차집합 :  {'강호동', '정수민'}




### 문제 7. 집합 조작하기

fruits = {'apple', 'banana', 'orange'}


다음 작업을 수행하는 코드를 작성하시오:


# 1. 'grape'를 추가하시오

fruits = set(['apple','banana','orange'])
fruits.update(['grape'])

# 2. 'melon', 'kiwi', 'mango'를 한 번에 추가하시오

fruits = set(['apple','banana','orange','grape'])
fruits.add('grape')

# 3. 'banana'를 제거하시오
fruits = set(['apple', 'banana', 'orange','grape','melon', 'kiwi', 'mango'])
fruits.remove('banana')

# 4. 집합에 'strawberry'가 있는지 확인하시오
'strawberry'in fruits

# 5. 모든 과일을 제거하고 빈 집합으로 만드시오
fruits = set(['apple', 'banana', 'orange','grape','melon', 'kiwi', 'mango'])
fruits.clear()



### 보너스 문제(선택). 영어-한글 사전 만들기 -> while
간단한 영한사전 프로그램을 만드시오.

# 최소 10개 이상의 단어를 포함한 영한사전 딕셔너리 생성
# 사용자가 영어 단어를 입력하면 한글 뜻을 출력
# 없는 단어면 "사전에 없는 단어입니다" 출력
# 'quit' 입력 시 종료
```
