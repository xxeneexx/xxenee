````python

### 문제 1-1. 딕셔너리 기본 개념
다음 질문에 답하시오.

1. 딕셔너리는 무엇으로 구성되어 있는가?

Key 와 Value 의 쌍으로 구성되어 있습니다.

2. 딕셔너리와 리스트의 가장 큰 차이점은 무엇인가?

리스트는 인덱스(0, 1, 2...)로 접근하고, 딕셔너리는 Key로 접근합니다.
리스트는 순서가 있지만, 딕셔너리는 Key-Value 매핑 관계가 중요합니다.

3. 다음 중 딕셔너리의 Key로 사용할 수 없는 것은? (이유도 함께 설명)
   - a) 숫자 (1, 2, 3)
   - b) 문자열 ("name", "age")
   - c) 리스트 ([1, 2, 3])
   - d) 튜플 ((1, 2, 3))

c) 리스트 ([1, 2, 3])
이유: Key는 변경 불가능한(immutable) 자료형이어야 합니다. 리스트는 변경 가능한(mutable) 자료형이므로 Key로 사용할 수 없습니다.
숫자, 문자열, 튜플은 모두 불변 자료형이므로 Key로 사용 가능합니다.

### 문제 1-2. 코드 읽기
다음 코드의 실행하시오

student = {'name': '김철수', 'age': 20, 'major': '컴퓨터공학'}
print(student['name'])
print(student['age'])
---

```
김철수
20

### 문제 2-1. 딕셔너리 생성 및 조작
다음 조건에 맞는 딕셔너리를 생성하고 조작하시오.


# 1. 다음 정보를 담는 딕셔너리를 만드시오
#    - 이름: 본인 이름
#    - 나이: 본인 나이
#    - 취미: 본인 취미 (리스트로 여러 개)
#    - 거주지: 본인 거주지

my_info = {
    '이름': '이순재',
    '나이': 30,
    '취미': ['독서', '영화감상', '운동'],
    '거주지': '포항'
}

# 2. 위 딕셔너리에 '직업' 항목을 추가하시오

my_info['직업'] = '직장인'

# 3. '취미' 항목에 새로운 취미를 추가하시오

my_info['취미'].append('여행')

# 4. '나이' 항목의 값을 1 증가시키시오

my_info['나이'] += 1

# 5. 모든 key와 value를 출력하시오

for key, value in my_info.items():
    print(f"{key}: {value}")
```
**실행 결과:**
```
이름: 이순재
나이: 31
취미: ['독서', '영화감상', '운동', '여행']
거주지: 포항
직업: 직장인

### 문제 2-2. 딕셔너리 함수 활용
다음 딕셔너리가 주어졌을 때, 각 요구사항을 수행하시오.


fruits = {
    'apple': 1000,
    'banana': 1500,
    'orange': 2000,
    'grape': 3000
}

# 1. 모든 과일 이름(key)을 리스트로 출력하시오

print(list(fruits.keys()))

# 2. 모든 가격(value)을 리스트로 출력하시오

print(list(fruits.values()))

# 3. 'banana'의 가격을 출력하시오

print(fruits['banana'])

# 4. 'melon'이 딕셔너리에 있는지 확인하시오

print('melon' in fruits)

# 5. 'grape'를 pop() 함수를 사용하여 삭제하고, 삭제된 값을 출력하시오

deleted_value = fruits.pop('grape')
print(deleted_value)
---
**실행 결과:**
```
['apple', 'banana', 'orange', 'grape']
[1000, 1500, 2000, 3000]
1500
False
3000

### 문제 3-1. 성적 관리 프로그램
학생들의 성적을 관리하는 프로그램을 작성하시오.

# 다음 학생들의 성적 정보를 딕셔너리로 저장하시오
# 학생1: 이름=김민수, 국어=85, 영어=90, 수학=78
# 학생2: 이름=이영희, 국어=92, 영어=88, 수학=95
# 학생3: 이름=박철수, 국어=78, 영어=85, 수학=82

# 요구사항:
# 1. 각 학생의 총점과 평균을 계산하여 딕셔너리에 추가
# 2. 각 학생의 정보를 보기 좋게 출력
# 3. 평균이 가장 높은 학생 찾기


**출력 예시:**

=== 학생 성적 관리 ===
이름: 김민수
국어: 85점, 영어: 90점, 수학: 78점
총점: 253점, 평균: 84.33점
---
이름: 이영희
국어: 92점, 영어: 88점, 수학: 95점
총점: 275점, 평균: 91.67점
---
이름: 박철수
국어: 78점, 영어: 85점, 수학: 82점
총점: 245점, 평균: 81.67점
---
최고 성적 학생: 이영희 (평균: 91.67점)

# 학생 정보 딕셔너리 생성
students = [
    {'이름': '김민수', '국어': 85, '영어': 90, '수학': 78},
    {'이름': '이영희', '국어': 92, '영어': 88, '수학': 95},
    {'이름': '박철수', '국어': 78, '영어': 85, '수학': 82}
]

# 1. 각 학생의 총점과 평균 계산
for student in students:
    total = student['국어'] + student['영어'] + student['수학']
    average = total / 3
    student['총점'] = total
    student['평균'] = average

# 2. 각 학생의 정보 출력
print("=== 학생 성적 관리 ===")
for student in students:
    print(f"이름: {student['이름']}")
    print(f"국어: {student['국어']}점, 영어: {student['영어']}점, 수학: {student['수학']}점")
    print(f"총점: {student['총점']}점, 평균: {student['평균']:.2f}점")
    print("---")

# 3. 평균이 가장 높은 학생 찾기
top_student = max(students, key=lambda x: x['평균'])
print(f"최고 성적 학생: {top_student['이름']} (평균: {top_student['평균']:.2f}점)")

### 문제 3-2. 단어 카운터
문자열에서 각 단어가 몇 번 등장하는지 세는 프로그램을 작성하시오.

text = "python is great python is fun python is powerful"

# 요구사항:
# 1. 문자열을 공백으로 분리하여 단어 리스트 만들기
# 2. 딕셔너리를 사용하여 각 단어의 등장 횟수 세기
# 3. 결과를 출력

# 힌트: 
# - 딕셔너리에 key가 없을 때 get() 함수의 default 값 활용
# - 또는 'word' in dict_name 으로 확인


**출력 예시:**

python: 3번
is: 3번
great: 1번
fun: 1번
powerful: 1번

text = "python is great python is fun python is powerful"

# 1. 문자열을 공백으로 분리
words = text.split()

# 2. 딕셔너리로 단어 횟수 세기
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1

# 3. 결과 출력
for word, count in word_count.items():
    print(f"{word}: {count}번")
```

**실행 결과:**
```
python: 3번
is: 3번
great: 1번
fun: 1번
powerful: 1번

### 문제 4-1. 중첩 딕셔너리
다음과 같은 중첩 딕셔너리를 다루는 문제를 해결하시오.


school = {
    '1반': {
        '김철수': {'국어': 85, '영어': 90, '수학': 78},
        '이영희': {'국어': 92, '영어': 88, '수학': 95}
    },
    '2반': {
        '박민수': {'국어': 88, '영어': 85, '수학': 90},
        '최지영': {'국어': 95, '영어': 92, '수학': 88}
    }
}

# 요구사항:
# 1. '1반' 김철수의 수학 점수를 출력하시오
# 2. '2반' 박민수의 평균 점수를 계산하여 출력하시오
# 3. 전교에서 국어 점수가 가장 높은 학생을 찾으시오
# 4. 각 반의 평균 점수를 계산하시오

school = {
    '1반': {
        '김철수': {'국어': 85, '영어': 90, '수학': 78},
        '이영희': {'국어': 92, '영어': 88, '수학': 95}
    },
    '2반': {
        '박민수': {'국어': 88, '영어': 85, '수학': 90},
        '최지영': {'국어': 95, '영어': 92, '수학': 88}
    }
}

# 1. '1반' 김철수의 수학 점수 출력
print(f"김철수의 수학 점수: {school['1반']['김철수']['수학']}점")

# 2. '2반' 박민수의 평균 점수 계산
park_scores = school['2반']['박민수']
park_avg = sum(park_scores.values()) / len(park_scores)
print(f"박민수의 평균 점수: {park_avg:.2f}점")

# 3. 전교에서 국어 점수가 가장 높은 학생 찾기
max_korean = 0
top_student = ""
for class_name, students in school.items():
    for student_name, scores in students.items():
        if scores['국어'] > max_korean:
            max_korean = scores['국어']
            top_student = student_name

print(f"국어 최고 점수 학생: {top_student} ({max_korean}점)")

# 4. 각 반의 평균 점수 계산
for class_name, students in school.items():
    total = 0
    count = 0
    for student_name, scores in students.items():
        total += sum(scores.values())
        count += len(scores)
    class_avg = total / count
    print(f"{class_name} 평균: {class_avg:.2f}점")

### 문제 5. 다음 코드의 실행 결과를 예상하시오.

# 1)
s1 = set("Hello")
print(s1)

# 2)
s2 = set([1, 2, 3, 2, 1, 3, 4])
print(s2)

# 3)
s3 = {}
print(type(s3))

# 1)
s1 = set("Hello")
print(s1)
# 예상 결과: {'H', 'e', 'l', 'o'} (순서는 무작위, 중복 제거됨)

# 2)
s2 = set([1, 2, 3, 2, 1, 3, 4])
print(s2)
# 예상 결과: {1, 2, 3, 4} (중복 제거됨)

# 3)
s3 = {}
print(type(s3))
# 예상 결과: <class 'dict'> (빈 중괄호는 딕셔너리입니다!)


### 문제 6. 두 반의 학생 명단 비교

A반과 B반 학생 명단이 있습니다.


a_class = {'김철수', '이영희', '박민수', '최지영', '홍길동'}
b_class = {'박민수', '최지영', '정수민', '강호동', '이영희'}

다음 질문에 답하시오:


# 1. 두 반 모두에 있는 학생 (교집합)

# 2. 두 반을 합친 전체 학생 (합집합)

# 3. A반에만 있는 학생 (차집합)

# 4. B반에만 있는 학생 (차집합)

a_class = {'김철수', '이영희', '박민수', '최지영', '홍길동'}
b_class = {'박민수', '최지영', '정수민', '강호동', '이영희'}

# 1. 두 반 모두에 있는 학생 (교집합)
print("교집합:", a_class & b_class)

# 2. 두 반을 합친 전체 학생 (합집합)
print("합집합:", a_class | b_class)

# 3. A반에만 있는 학생 (차집합)
print("A반에만:", a_class - b_class)

# 4. B반에만 있는 학생 (차집합)
print("B반에만:", b_class - a_class)


### 문제 7. 집합 조작하기

fruits = {'apple', 'banana', 'orange'}


다음 작업을 수행하는 코드를 작성하시오:


# 1. 'grape'를 추가하시오

# 2. 'melon', 'kiwi', 'mango'를 한 번에 추가하시오

# 3. 'banana'를 제거하시오

# 4. 집합에 'strawberry'가 있는지 확인하시오

# 5. 모든 과일을 제거하고 빈 집합으로 만드시오


fruits = {'apple', 'banana', 'orange'}

# 1. 'grape' 추가
fruits.add('grape')

# 2. 'melon', 'kiwi', 'mango' 한 번에 추가
fruits.update(['melon', 'kiwi', 'mango'])

# 3. 'banana' 제거
fruits.remove('banana')

# 4. 'strawberry'가 있는지 확인
print('strawberry' in fruits)

# 5. 모든 과일 제거
fruits.clear()
print(fruits)

### 보너스 문제(선택). 영어-한글 사전 만들기
간단한 영한사전 프로그램을 만드시오.

# 최소 10개 이상의 단어를 포함한 영한사전 딕셔너리 생성
# 사용자가 영어 단어를 입력하면 한글 뜻을 출력
# 없는 단어면 "사전에 없는 단어입니다" 출력
# 'quit' 입력 시 종료

# 영한사전 딕셔너리
eng_kor_dict = {
    'apple': '사과',
    'banana': '바나나',
    'computer': '컴퓨터',
    'dictionary': '사전',
    'student': '학생',
    'teacher': '선생님',
    'book': '책',
    'school': '학교',
    'house': '집',
    'car': '자동차',
    'python': '파이썬'
}

print("=== 영한사전 ===")
print("'quit'을 입력하면 종료됩니다.")

while True:
    word = input("\n영어 단어를 입력하세요: ").lower()
    
    if word == 'quit':
        print("프로그램을 종료합니다.")
        break
    
    if word in eng_kor_dict:
        print(f"{word}: {eng_kor_dict[word]}")
    else:
        print("사전에 없는 단어입니다.")
````
