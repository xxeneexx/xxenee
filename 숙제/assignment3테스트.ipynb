```python

### 문제 3-2. 단어 카운터
#문자열에서 각 단어가 몇 번 등장하는지 세는 프로그램을 작성하시오.

text = "python is great python is fun python is powerful"

# 요구사항:
# 1. 문자열을 공백으로 분리하여 단어 리스트 만들기
# 2. 딕셔너리를 사용하여 각 단어의 등장 횟수 세기
# 3. 결과를 출력

# 힌트: 
# - 딕셔너리에 key가 없을 때 get() 함수의 default 값 활용
# - 또는 'word' in dict_name 으로 확인


#**출력 예시:**

#python: 3번
#is: 3번
#great: 1번
#fun: 1번
#powerful: 1번
```

```python
print("=== 학생 성적 관리 ===")

# 학생별 점수 딕셔너리
grade = {
    "김민수": {"국어":85, "영어":90, "수학":78},
    "이영희": {"국어":92, "영어":88, "수학":95},
    "박철수": {"국어":78, "영어":85, "수학":82}
}

# 최고 평균 저장용 변수
top_student = ""
top_avg = 0

# 김민수
total = grade["김민수"]["국어"] + grade["김민수"]["영어"] + grade["김민수"]["수학"]
avg = total / 3
print("이름: 김민수")
print("국어:", grade["김민수"]["국어"], "점, 영어:", grade["김민수"]["영어"], "점, 수학:", grade["김민수"]["수학"], "점")
print("총점:", total, "점, 평균:", "{:.2f}".format(avg), "점")
print("---")
if avg > top_avg:
    top_avg = avg
    top_student = "김민수"

# 이영희
total = grade["이영희"]["국어"] + grade["이영희"]["영어"] + grade["이영희"]["수학"]
avg = total / 3
print("이름: 이영희")
print("국어:", grade["이영희"]["국어"], "점, 영어:", grade["이영희"]["영어"], "점, 수학:", grade["이영희"]["수학"], "점")
print("총점:", total, "점, 평균:", "{:.2f}".format(avg), "점")
print("---")
if avg > top_avg:
    top_avg = avg
    top_student = "이영희"

# 박철수
total = grade["박철수"]["국어"] + grade["박철수"]["영어"] + grade["박철수"]["수학"]
avg = total / 3
print("이름: 박철수")
print("국어:", grade["박철수"]["국어"], "점, 영어:", grade["박철수"]["영어"], "점, 수학:", grade["박철수"]["수학"], "점")
print("총점:", total, "점, 평균:", "{:.2f}".format(avg), "점")
print("---")
if avg > top_avg:
    top_avg = avg
    top_student = "박철수"

# 최고 성적 학생 출력
print("최고 성적 학생:", top_student, "(평균:", "{:.2f}".format(top_avg), "점)")
```

```python
text = "python is great python is fun python is powerful"

# 1. 공백으로 분리하여 단어 리스트 만들기
words = text.split(" ")

# 2. 단어 등장 횟수 세기
word_count = {}

# 단어 하나씩 확인하며 개수 세기
for w in words:
    # key가 없으면 0, 있으면 현재 값 +1
    word_count[w] = word_count.get(w, 0) + 1

# 3. 결과 출력
print("=== 단어 등장 횟수 ===")
print("python:", word_count["python"], "번")
print("is:", word_count["is"], "번")
print("great:", word_count["great"], "번")
print("fun:", word_count["fun"], "번")
print("powerful:", word_count["powerful"], "번")

```
