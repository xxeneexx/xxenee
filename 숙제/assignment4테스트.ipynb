```python
### 문제 7-1: 택시 요금 계산기

#거리(km)와 시간대를 입력받아 요금을 계산하세요.

# - 기본요금: 3000원 (2km까지)
# - 추가요금: 2km 초과시 1km당 1000원
# - 심야할증(22시~04시): 총 요금의 20% 추가
distance = float(input("거리를 입력하세요 (km): "))
hour = int(input("현재 시간을 입력하세요 (0-23): "))
#   **예상 출력:**
#   거리를 입력하세요 (km): 5
#   현재 시간을 입력하세요 (0-23): 23
#   요금: 7200원


# 기본 요금 계산
if distance <= 2:
    fare = 3000
else:
    fare = 3000 + (distance - 2) * 1000

# 심야 할증
if hour >= 22 or hour <= 4:
    fare = fare + (fare * 0.2)

print(f"요금: {int(fare)}원") 

# 출력하기

if distance < 2:
    print("기본요금: 3000원")
elif distance > 2 or hour >= 23:
    print("거리를 입력하세요 (km): 5")
    print("현재 시간을 입력하세요 23") 
    print("요금: 7200 원")
```

```python

### 문제 7-2: 로그인 시스템

#아이디와 비밀번호를 입력받아 로그인을 처리하세요.

# - 올바른 아이디: "admin"
# - 올바른 비밀번호: "1234"
# - 아이디가 틀리면: "아이디를 확인하세요"
# - 비밀번호가 틀리면: "비밀번호를 확인하세요"
# - 둘 다 맞으면: "로그인 성공!"
user_id = input("아이디: ")
password = input("비밀번호: ")

if user_id == "admin" and password == "1234":
    print("로그인 성공")
elif user_id != "admin":
    print("아이디를 확인하세요")
elif password != "1234":
    print("비밀번호를 확인하세요")
```
