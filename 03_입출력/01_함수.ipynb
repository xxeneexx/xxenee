```python
# 함수란 무엇인가?
#중복되는 반복되는 것을 피할때 (반복되는 코드를 줄여줌,같은 계산, 같은 작업을 여러 번 쓸 필요 없음)
#코드를 깔끔하게 코드를 블록 단위로 나눠서 보기 좋게 함
#x는 인풋 y는 결과값  들어가는거에 따라 결과값이 달라지는게 함수
#믹서기, 자판기 

#같은 일을 반복해서 쓰지 않도록, 미리 만들어 놓은 코드 덩어리
#입력 → 처리 → 출력 구조를 가진 작업 단위
#즉, **입력(x)**이 들어가면, **결과값(y)**이 나오는 일종의 '기계'예요.


#파이썬 함수의 구조
#def 정의하라 (디파인) 파이썬의 예약어

# def 함수이름(매개변수):  매개변수는 그냥 변수 호출한 사람
#     수행할_문장1
#     수행할_문장2
#     ...
                 #수행은 함수 밖에서 해야함 같은줄에쓰면 수행중인거
def add(a, b):   #:는 뒤에 수용할 문장이 옴. a,b는 매개변수
    return a + b


result = add(3,4)  #print(add(3,4))  = add(3,4) , #add는 함수를 호출하는것이다 
print(result)            
                  # a = 3 a는 매개변수 3은 인수 
```

```python
# 입력값이 몇 개가 될지 모를 때는 어떻게 해야 할까?

#args :인수, 입력값의 약자 , 함수에 몇 개의 입력값이 들어올지 모를 때 사용하는 매개변수
# *을 붙이면 여러 입력값을 하나로 묶어서 튜플로

# def 함수이름(*args):  한개가될수도있고 여러개가 될지 모를때 * 기로 표기
#     수행할문장1
#     수행할문장2
#      ...

def add_many(*args): # args = [1,2,3,4,5]
    result = 0   # 초기화 하는값 0 
    for i in args:  #튜플 리스트 여러가지개가 되는거가 여기자리
        result = result + i
    return result

a = add_many(1,2,3,4,5)  #add를 호출하면 맨위 def 부터 실행해라 이얘기고 실행이 끝나면 종료되는데 그때 a = 15가 되는데 print(a)하면됨
print(a)

```
