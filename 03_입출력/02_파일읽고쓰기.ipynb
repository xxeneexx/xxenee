```python
f =open("data2", "w")  #r은 리드 w는 라이트의쓰기  #컴퓨터 메모장에 글쓰고 해당 폴더에 넣어놓으면됨
f.close()

# 파일객체 = open(파일이름, 파일열기모드)

# r - 읽기모드: 파일을 읽기만 할때 
# w - 쓰기모드: 파일의 내용을 쓸때
# a - 추가모드: 파일의 마지막에 내용을 추가할때 
```

```python
f = open("./data/새파일.text", "w") #경로  .은 현재디렉토리 상대경로

for i in range(1,11):
    data = f"{i}번째 줄입니다.\n" 
    f.write(data)

f.close()
```

```python
f = open("./data/새파일.txt", "r") #오픈이라는 매써드로 객체를..
line = f.readline() #리드라인은 한줄씩 읽는거. 라인이라고 객체둠
print(line)
f.close()

```

```python
f = open("data/새파일.text", "r")
while True:   #와일로함 여러번 반복해야해서
    line = f.readline()
    if not line: break
    print(line)

f.close()
```

```python
# readlines() 함수 이용하기
f = open("data/새파일.text", "r")
lines = f.readlines()  #여러줄을 읽는거 
print(lines)
f.close

for line in lines:
    print(line)
f.close()


print(type(lines))
print(len(lines))
```

```python
#read 함수 사용하기

f = open("data/새파일.text", "r")
data = f.read()
print(data)
f.close()


```

```python
# 파일에 새로운 내용 추가하기
f = open("../새파일.txt", "a")  #폴더안이아니라 밖으로 새파일.text 꺼내서 저장해놨음

for i in range(11,21):
    f.write(f"{i}번째 줄입니다.\n")

f.close()
```

```python
# with 문과 함께 사용하기.  #as ~로써 별칭을 쓸때  만들어진거의 이름을 f로 하겠다
with open("data3.txt", "w") as f:
    f.write("Life is too short, you need Python") #whit문은 close 따로 해줄필요없음
```
