```python
#딕셔너리는 어떻게 만들까? #딕셔너리는 사전같은거임
dic={
"name" : "json",  #네임이 키 , 제이슨이 벨류
"phone" : "010-1234-1234", 
"birth" : "00-01-01"
}
```

```python
#딕셔너리 쌍 추가,삭제하기  
#딕셔너리라는 객체
dic["hobby"]="running"  # 추가
dic
```

```python

dic = {}  # 빈 딕셔너리 생성
dic[3] = [1,2,3]  # 값 추가
```

```python
#딕셔너리 요소 삭제하기
del dic[3]
```

```python
del dic["hobby"] #호비를 키라 부름
dic
```

```python
#딕셔너리 요소 변경하기 
dic["name"]="python"   
dic
```

```python
#딕셔너리에서 key를 사용해서 값(Value) 얻기
print(dic["name"])
```

```python
#딕셔너리에서 key를 사용해서 값(Value) 얻기
print(dic["name"])
print(dic["phone"])
print(dic["birth"])
print(dic["hobby"])
dic
```

```python
#딕셔너리 관련 함수 
#Keys 리스트 만들기 - keys
#dic는 하나의 객체 자동차 같은 존재. 저거 자체로만으로도 리스트로 변환이 가능
dic={
"name" : "json", 
"phone" : "010-1234-1234", 
"birth" : "00-01-01"
}
dic.keys() #키즈는 값을 가져오는 함수
list(dic.keys())  #리스트로 변환 리스트타입의 데이로 반환됨
```

```python
#Values 리스트 만들기 - values
dic.values()
list(dic.values()) #list라는 내장함수에 나머지를 그 안에 넣은거
```

```python
# Key, Value 쌍으로 얻기 -items 
li=dic.items()
list(li)
```

```python
#Key, Value 쌍 모두 지우기 - clear  (함수는 매써드)
dic.clear()
dic 
```

```python
#Key로 Value 얻기 - get (get이라는 매써드를 이용해서 얻는다) 
dic={
"name" : "json",  
"phone" : "010-1234-1234", 
"birth" : "00-01-01"
} 
# dic["name"]
dic.get("name")

# dic["name"]
print(dic.get("name"))
print(dic.get("phone"))
print(dic.get("birth"))
```

```python
#해당 Key가 딕셔너리 안에 있는지 조사하기 - 

dic={
"name" : "json",  
"phone" : "010-1234-1234", 
"birth" : "00-01-01"
}

print("name" in dic)
print("email" in dic)

```

```python
#Key 로 Value 얻기 - pop

dic={
"name" : "json",  
"phone" : "010-1234-1234", 
"birth" : "00-01-01"
}

phone = dic.pop('phone')
print(phone)

dic
```
