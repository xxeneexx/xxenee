```python
#교집합, 합집합,차집합 구하기 
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])

## 교집합 구하기 
print("교집합 : ", s1 & s2)

##합집합 구하기 
print("합집합 : ", s1 | s2) # 벽같이생긴거 올연산자

##차집합 구하기
print("차집합 : ", s1 -s2)
print("차집합 : ", s1 -s2)
```

```python
#집합 자료형 관련 함수
s1 = set([1,2,3]) #s1은 집합형태를가지고있는 객체
s1

##값1개 추가하기 -add
s1.add(4)  #.을쓰는거는 매써드를쓰는거임 s1의 속성안에 있는걸 쓰는거임
s1
```

```python

## 값 여러개 추가하기 - update
s1 = set([1,2,3])
s1.update([4,5,6])

{1,2,3,4,5,6}
```

```python
## 특정 값 제거하기 -remove 

s1 = set([1,2,3])
s1.remove(2)

{1,3}
```

```python

#모든 값 제거하기
s1 = set([1,2,3])
s1.clear()

set() #모든것이비어있는거 
```
