```python
#자료형의 참과 거짓
print(bool("python")) #비교연산이아니라 그냥 컴퓨터는 들어간거에대한 참인지 거짓인지에 대한걸로 판단한다
print(bool(" "))

print(bool("python"))   # True, 문자열이 있으니까
print(bool(" "))        # True, 공백 문자열도 비어있지 않으니까
print(bool([1,2,3]))    # True, 리스트에 값이 있으니까
print(bool([]))         # False, 빈 리스트니까
print(bool([{}]))       # True, 리스트 안에 빈 딕셔너리라도 리스트는 비어있지 않음
print(bool(1))          # True, 0이 아니니까
print(bool(0))          # False
print(bool(None))       # False


```

```python
# and 연산자, or 연산자 

print(True and True)
print(True and False)
print(False and True)
print(False and False)

print(True or True)
print(True or False)
print(False or True)
print(False or False)
```
