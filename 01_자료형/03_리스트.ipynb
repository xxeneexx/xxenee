```python

# a = [1, 4, 3, 2]

a.sort()                # 오름차순(Ascending)
print(a)                # [1, 2, 3, 4]

a.sort(reverse=True)    # 내림차순(Descending)
print(a)                # [4, 3, 2, 1]

#reverse= 뒤집겠다 
#bool 참과 거짓
```

```python
#인덱스 반환하는 함수 - index 
# 리스트 a에서 값 1이 있는 위치의 인덱스를 반환
a=[1,2,3]
a.index(1)  # 값 1의 인덱스 반환 → 0
a.index(2)  # 값 2의 인덱스 반환 → 1
a.index(3)  # 값 3의 인덱스 반환 → 2

```

```python
#리스트에 요소 삽입
#insert(삽입할 위치, 삽입할 값)
a=[1,2,3]
a.insert(0,4)
a
```

```python
#리스트요소 제거 -remove
# 리스트에서 "첫 번째로 나오는 3" 제거
a=[1,2,3,1,2,3]
a.remove(3)
a
```

```python
#리스트 요소 꺼내기 -pop (기본값이 -1일임 그래서 뒤에서부터 3이 없어지고 값이 1,2가 나옴)
# 리스트 a의 마지막 요소를 제거하고 그 값을 반환
#pop() → 마지막 요소 제거
#pop(-1) → 마지막 요소 제거 (기본값) ->  pop()=pop(-1)
#pop(0) → 첫 번째 요소 제거
#제거된 값은 반환됨

a=[1,2,3]
a.pop()
a
```

```python
#리스트에 포함된 요소 x의 개수 세기 -count
#count(값) → 리스트 안에서 그 값이 몇 개 있는지 세는 함수
a=[1,2,3,1]
a.count(2) #리스트 안에서 2가 몇 번 등장?
```
