```python
# 클래스는 왜 필요한가? #함수로 대체됨(c언어는 클래스가 없슴). #next js 많이씀..? 
# 설계도
#매써드 = 함수 속성 = 변수 

#계산기만들기
result1 = 0

def add1(num):
    global result1  
    result1 += num  # 결괏값(result)에 입력값(num) 더하기
    return result1  # 결괏값 리턴

print(add1(3))
print(add1(4))
print(add1(5))
```

```python
class Calculator:   #클래스는 예약어 클래스는 클래스 다음에 
    #적는변수는 대문자로시작해야함
    def __init__(self):#self 는 객체 자기 자신 가리키는거임 
        self.result = 0

    def add(self, num):
        self.result  = self.result + num #self.result += num
        return self.result

cal1 = Calculator()  #cal1 객체 cal2객체
cal2 = Calculator()

(cal1.add(3))
(cal2.add(4))

print(cal1.add(3))
print(cal2.add(4))
```

```python
#더하기 기능만들기

class FourCal:
    def setdata(self, first, second): #생성자에의해 만들어진 객체 self
        self.first = first 
        self.second = second

    def add(self): #매써드는 add 함수
        result = self.first + self.second # a에 first a에 second
        return result

a = FourCal() #a라는 변수는 class에의해 만들어졌음
type(a)
a.setdata(3, 4)
print(a.first)
print(a.second)

a.add()   #()하면 함수 안하면 함수아닌거 .append 클래스의 어펜드 .점이 그런의미임
```

```python
# 곱하기, 빼기, 나누기 기능 만들기
# nul() - 곱하기
# sub() - 빼기
# div() - 나누기

### 계산기 2개를 만들어서 (cal1 , cal2)
### cal1(12, 20)
### cal(20, 5)
#각각의 곱하기 빼기, 나누기를 모두 만드시오


class FourCal:
    def setdata(self, first, second): 
        self.first = first 
        self.second = second

    def add(self): 
        result = self.first + self.second 
        return result

    def mul(self): 
        result = self.first * self.second 
        return result

    def sub(self): 
        result = self.first - self.second 
        return result

    def div(self): 
        result = self.first / self.second 
        return result

cal1 = FourCal()
cal1.setdata(10, 20)
print(cal1.add())
print(cal1.mul())
print(cal1.sub())
print(cal1.div())

cal2 = FourCal() # cal2 변수이지만 객체이기도하다
cal2.setdata(20, 5)
print(cal2.add())
print(cal2.mul())
print(cal2.sub())
print(cal2.div()) #.div 점쓰고 뒤에쓰는건 매써드 ()괄호를통해서 함수 실행시킴

```
