```python
# all() 요소가 모두 참이면 -> 참
all([1,2,3])
```

```python
# any() 요소중에서 참이 하나라도 있으면 ->
any([0,1,2,""])
```

```python
#dir -객체가 지닌 변수나 함수를 보여주는 함수
dir([1,2,3]) #dir안에있는 객체가 리스트형식으로되어있는거
#여러가지 매써들이 기억한날때 dir 쓰면됨
```

```python
# id -> 객체를 입력받아 객체의 고유 주소값을 반환하는 함수
a = 3
b = 4
#a에 3이 담겨진게아니라 어떤변수에 3이 있는데 그걸 가리키는 주소가 a임

print(id(a))
print(id(b))
```

```python
# map(f, iterable)은 각 요소를 f함수에 적용한 결과를 보여주는 함수

def tow_times(x):
    return x*2

list(map(tow_times, [1,2,3,4]))
```
